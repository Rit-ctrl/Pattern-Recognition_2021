# Q2. Find and plot the decision boundary between class ω1 and ω2. Assume    P(ω1) =0.3; P(ω2)=0.7

ω1 = [1,-1; 2,-5; 3,-6; 4,-10; 5,-12; 6,-15]

ω2 = [-1,1; -2,5; -3,6; -4,10, -5,12; -6, 15]


```python
w1 = [[1,-1],[2,-5],[3,-6],[4,-10],[5,-12],[6,-15]]
w2 = [[-1,1],[-2,5],[-3,6],[-4,10],[-5,12],[-6,15]]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
w1=pd.DataFrame(data=w1,columns=['x','y'])
w2=pd.DataFrame(data=w2,columns=['x','y'])

#converting to pandas dataframe for ease of access and plotting
```

```python
print("Class w1 \n",w1)
print("\n")
print("Class w2 \n",w2)
print("\n")

plt.scatter(x=w1['x'],y=w1['y'],label="Class w1")
plt.scatter(x=w2['x'],y=w2['y'],label="Class w2")
plt.legend()
plt.title("2 classes ")
```

## Calc mean and covariance

```python
#mean
w1_mean=w1.mean().to_numpy()
w2_mean=w2.mean().to_numpy()

print(w1_mean)
print(w2_mean)
```

```python
#covariance

w1_cov=w1.cov().to_numpy()
w2_cov=w2.cov().to_numpy()

print(w1_cov)
print(w2_cov)
```

## Defining discriminant function

g.JPG

```python
def gi(x,mean,cov,p_w):
  t1=-0.5 *(np.subtract(x,mean)).T @ np.linalg.inv(cov) @ np.subtract(x,mean)
  t2=-0.5 *np.log(np.linalg.det(cov))
  t3=np.log(p_w)
  return t1+t2+t3
```

```python
from sympy import Matrix, solve, symbols
```

```python
x,y=symbols('x y')
```

```python
M=Matrix([x,y])
```

```python
M
```

```python
g1=gi(M,w1_mean.reshape(2,1),w1_cov,0.3)
g2=gi(M,w2_mean.reshape(2,1),w2_cov,0.7)
```

```python
g1
```

```python
g=np.subtract(g1[0][0],g2[0][0])
```
