## Function for calculating P(X/wi)

```python
def cal_prob(dim,det,X,mean,inverse):
    t1=1/( (np.pi**(dim/2)) * (det**0.5))
    t2=(np.transpose(np.subtract(X,mean))) @ inverse @ (np.subtract(X,mean))
    return t1 * np.exp(-0.5 * t2)
```

## Finding probability


```python
def predict(X):
    p_fm=cal_prob(128,det_cov_fm,X,mean_fm,inv_cov_fm)
    p_m=cal_prob(128,det_cov_m,X,mean_m,inv_cov_m)
    pro_fm=p_fm*pw_fm
    pro_m=p_m*pw_m
    class_pred=max(pro_fm,pro_m)
    print(pro_fm,pro_m)
    
    if(class_pred==pro_fm):
        return "female"
    else:
        return "male"
```

## Testing female

```python
true_class_fm=[]
results_fm=[]
for i in range(5):
    results_fm.append(predict(df_fm_test[df_fm_test.columns[:]].iloc[i]))
    true_class_fm.append("female")
```

## Calculating the accuracy 

```python
def accuracy(pred,true_class):
    true=0
    total=len(pred)
    for i in range(total):
        if pred[i] == true_class[i]:
            true+=1
    
    return true/total
```

```python
print("accuracy for male test cases : ",accuracy(results_m,true_class_m))
```
